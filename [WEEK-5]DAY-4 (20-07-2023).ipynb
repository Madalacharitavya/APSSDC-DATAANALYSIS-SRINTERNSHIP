### Joint plot

* Combination of two plots. By default it have scatter plot,histogram
* sns.jointplot()

```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
help(sns.jointplot)
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris)
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,color='g')
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,kind='hex',color='orange')
```

```python
# draw a joint plot by taking kind as kde
```

```python
sns.jointplot(x='petal_width',y='sepal_width',data=iris,kind='kde',color='orange')
```

### Pairplot

* It displays multiple plots at a time in a single graph
* Pairwise Relationship
* Bydefault it return scatterplot
     * sns.pairplot()

```python
iris
```

```python
sns.pairplot(iris)
```

```python
help(sns.pairplot)
```

```python
sns.pairplot(iris,hue='species')
```

```python
# draw a pairplot with palette
```

```python
sns.pairplot(iris,hue='species',palette='husl')
```

```python
sns.pairplot(iris,hue='species',palette='dark')
```

```python
sns.pairplot(iris,hue='species',markers=['o','D','s'])
```

```python
sns.pairplot(iris,vars=["sepal_length","sepal_width"])
```

### Heatmap

* It is used to correlation between two columns in data

```python
help(sns.heatmap)
```

```python
t=sns.load_dataset("tips")
```

```python
t.head()
```

```python
t.corr()
```

```python
help(t.corr)
```

```python
sns.heatmap(t.corr())
```

```python
sns.heatmap(t.corr(),annot=True)
```

```python
sns.heatmap(t.corr(),annot=True,center=0,linewidth=5,linecolor='black')
```

```python
sns.heatmap(t.corr(),annot=True,cmap='Blues')
```

```python
sns.heatmap(t.corr(),annot=True,cmap='YlGnBu')
```
